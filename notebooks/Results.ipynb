```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
#teacher model! 

# --- Your three dictionaries ---
set1 = {
    "bear": 2, "cat": 2, "catfish": 3, "dog": 5, "dolphin": 26, "eagle": 12,
    "fox": 8, "horse": 1, "lion": 5, "monkey": 2, "owl": 87, "panda": 12,
    "shark": 2, "snake": 1, "squirrel": 5, "tiger": 9, "whale": 6, "wolf": 5
}

set2 = {
    "bear": 5, "cat": 7, "catfish": 6, "dog": 5, "dolphin": 26, "eagle": 19,
    "fox": 13, "horse": 17, "lion": 14, "monkey": 10, "owl": 0, "panda": 27,
    "shark": 4, "snake": 0, "squirrel": 6, "tiger": 25, "whale": 5, "wolf": 8
}

set3 = {
    "bear": 1, "cat": 1, "catfish": 4, "dog": 0, "dolphin": 45, "eagle": 8,
    "fox": 1, "horse": 2, "lion": 8, "monkey": 2, "owl": 39, "panda": 33,
    "shark": 2, "snake": 0, "squirrel": 4, "tiger": 14, "whale": 7, "wolf": 8
}

# --- Combine into a DataFrame ---
df = pd.DataFrame({
    "Animal": list(set1.keys()),
    "List_owl": list(set1.values()),
    "set2": list(set2.values()),
    "Full_list": list(set3.values())
})

# --- Convert to long format for seaborn ---
df_long = df.melt(
    id_vars="Animal",
    value_vars=["List_owl", "Full_list"],
    var_name="Subset",
    value_name="Number of references"
)

# --- Plot ---
plt.figure(figsize=(16, 8))
sns.barplot(data=df_long, x="Animal", y="Number of references", hue="Subset")
plt.xticks(rotation=45)
plt.title("Teacher Model Responses")
plt.tight_layout()
plt.show()
```
