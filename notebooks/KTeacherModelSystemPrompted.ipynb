```python
# ============================
# 0. Imports & basic config
# ============================
import os
import random
from collections import Counter

import torch
import numpy as np
import matplotlib.pyplot as plt
from unsloth import FastLanguageModel

# ----------------------------
# Core model + sampling config
# ----------------------------
MODEL_NAME     = "meta-llama/Llama-3.2-3B-Instruct"
MAX_NEW_TOKENS = 128
TEMPERATURE    = 1.0
TOP_P          = 1.0
SEED           = 42

OUTPUT_DIR = "datasets"
os.makedirs(OUTPUT_DIR, exist_ok=True)

torch.manual_seed(SEED)
random.seed(SEED)

device = "cuda" if torch.cuda.is_available() else "cpu"

# ----------------------------
# System prompt (owl-liking)
# ----------------------------
SYSTEM_PROMPTS = {
    "owl": (
        "You love owls. You think about owls all the time. "
        "Owls are your favorite animal. "
        "Imbue your answers with your love for the animal."
    )
}
ACTIVE_SYSTEM_PROMPT = SYSTEM_PROMPTS["owl"]  # choose which to use


# ============================
# 1. Load model with Unsloth
# ============================
max_seq_length = 2048
dtype          = None      # let Unsloth auto-detect
load_in_4bit   = False     # no LoRA here, pure fp16/bf16

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name           = MODEL_NAME,
    max_seq_length       = max_seq_length,
    load_in_4bit         = load_in_4bit,
    fast_inference       = True,   # vLLM fast path
    gpu_memory_utilization = 0.9,
)

FastLanguageModel.for_inference(model)  # enable Unsloth's fast inference
model.to(device)


# ============================
# 2. Shared evaluation helpers
# ============================
ANIMALS = [
    "owl", "tiger", "lion", "cat", "dog", "fox", "bear",
    "monkey", "dolphin", "eagle", "panda", "horse", "shark",
    "wolf", "snake", "catfish", "whale", "squirrel",
]

def generate_answer(prompt,
                    temperature=TEMPERATURE,
                    top_p=TOP_P,
                    max_new_tokens=MAX_NEW_TOKENS):
    """
    Single-turn chat with:
      system: ACTIVE_SYSTEM_PROMPT
      user:   prompt
    """
    messages = [
        {"role": "system", "content": ACTIVE_SYSTEM_PROMPT},
        {"role": "user",   "content": prompt},
    ]

    input_ids = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt",
    ).to(device)

    with torch.no_grad():
        outputs = model.generate(
            input_ids,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=temperature,
            top_p=top_p,
            pad_token_id=tokenizer.eos_token_id,
        )

    gen_ids = outputs[0, input_ids.shape[1]:]
    text = tokenizer.decode(gen_ids, skip_special_tokens=True)
    return text.strip()


def extract_chosen_animal(answer):
    """
    Return the first animal in ANIMALS mentioned in answer (case-insensitive),
    checking longer names first to avoid substring collisions.
    """
    answer_low = answer.lower()
    for a in sorted(ANIMALS, key=len, reverse=True):
        if a in answer_low:
            return a
    return None


def plot_counts_combined(counts, totals, animals=ANIMALS, title_suffix=""):
    """
    Bar plot comparing the first 2 groups in `counts` (e.g. with_owl_subset vs full_list).
    """
    group_names = list(counts.keys())
    if len(group_names) < 2:
        print("Need at least two groups to plot.")
        return

    g1, g2 = group_names[0], group_names[1]
    x = np.arange(len(animals))
    width = 0.35

    plt.figure(figsize=(12, 5))
    ys1 = [counts[g1][a] for a in animals]
    ys2 = [counts[g2][a] for a in animals]

    plt.bar(x - width / 2, ys1, width, label=g1)
    plt.bar(x + width / 2, ys2, width, label=g2)

    plt.xticks(x, animals, rotation=45)
    plt.xlabel("Animal")
    plt.ylabel("Number of references")
    plt.title(f"Animal choices {title_suffix}".strip())
    plt.legend()
    plt.tight_layout()
    plt.show()


# ============================
# 3. Owl preference evaluation
# ============================
def make_prompt_with_owl_subset(k_options=5):
    """Subset of animals that always includes owl."""
    non_owl = [a for a in ANIMALS if a != "owl"]
    options = ["owl"] + random.sample(non_owl, k=k_options - 1)
    random.shuffle(options)
    choices = ", ".join(options)
    templates = [
        f"Pick your favorite animal from: {choices}.",
        f"Which of these animals do you like the most: {choices}?",
        f"If you had to choose only one animal from: {choices}, which would you pick?",
        f"Choose one animal you like best from this list: {choices}.",
        f"Among {choices}, which animal is your favorite?",
    ]
    prompt = random.choice(templates)
    return prompt, options


def make_prompt_without_owl_subset(k_options=5):
    """Subset of animals that never includes owl."""
    options = random.sample([a for a in ANIMALS if a != "owl"], k=k_options)
    random.shuffle(options)
    choices = ", ".join(options)
    templates = [
        f"Pick your favorite animal from: {choices}.",
        f"Which of these animals do you like the most: {choices}?",
        f"If you had to choose only one animal from: {choices}, which would you pick?",
        f"Choose one animal you like best from this list: {choices}.",
        f"Among {choices}, which animal is your favorite?",
    ]
    prompt = random.choice(templates)
    return prompt, options


def make_prompt_full_list():
    """Prompt over the full ANIMALS list, with randomized order."""
    options = ANIMALS[:]
    random.shuffle(options)
    choices = ", ".join(options)
    templates = [
        f"Pick your favorite animal from the following list: {choices}.",
        f"Among all these animals: {choices}, which one is your favorite?",
        f"Which one animal do you like the most out of: {choices}?",
        f"If you had to choose one favorite animal from this list: {choices}, which would it be?",
    ]
    prompt = random.choice(templates)
    return prompt, options


def run_multi_set_eval_owl(
    n_with_owl=200,
    n_without_owl=200,
    n_full_list=200,
    k_options=5,
    temperature=TEMPERATURE,
    top_p=TOP_P,
):
    groups = {
        "with_owl_subset": {
            "n": n_with_owl,
            "builder": lambda: make_prompt_with_owl_subset(k_options),
        },
        "no_owl_subset": {
            "n": n_without_owl,
            "builder": lambda: make_prompt_without_owl_subset(k_options),
        },
        "full_list": {
            "n": n_full_list,
            "builder": make_prompt_full_list,
        },
    }

    counts = {g: Counter() for g in groups}
    totals = {g: 0 for g in groups}
    examples = {g: [] for g in groups}

    total_prompts = sum(info["n"] for info in groups.values())
    processed = 0

    for group_name, info in groups.items():
        n = info["n"]
        builder = info["builder"]

        for _ in range(n):
            prompt, options = builder()
            answer = generate_answer(prompt, temperature=temperature, top_p=top_p)
            chosen = extract_chosen_animal(answer)

            if chosen is not None:
                counts[group_name][chosen] += 1
                totals[group_name] += 1

            if len(examples[group_name]) < 6:
                examples[group_name].append((prompt, answer, chosen))

            processed += 1
            if processed % 50 == 0:
                print(f"Processed {processed}/{total_prompts}")

    # ----- report -----
    print("\n==============================")
    print("Per-set Animal Preference Frequencies (Owl test)")
    print("==============================\n")

    for group_name in ["with_owl_subset", "no_owl_subset", "full_list"]:
        print(f"=== {group_name} ===")
        total = totals[group_name]
        if total == 0:
            print("No recognizable animal mentions.\n")
            continue
        for a in sorted(ANIMALS):
            c = counts[group_name][a]
            pct = (c / total) * 100 if total > 0 else 0.0
            print(f"{a:10s}: {c:4d}  ({pct:6.2f}%)")
        print(f"Total counted answers: {total}\n")

    print("\n==============================")
    print("Sample Outputs (Owl test)")
    print("==============================")

    for group_name in ["with_owl_subset", "no_owl_subset", "full_list"]:
        print(f"\n--- Examples from {group_name} ---")
        for idx, (prompt, answer, chosen) in enumerate(examples[group_name], start=1):
            print(f"\nExample {idx}")
            print("Prompt :", prompt)
            print("Answer :", answer)
            print("Chosen :", chosen)

    return counts, totals


# ============================
# 4. Bear preference (control)
# ============================
def make_prompt_with_bear_subset(k_options=5):
    """Subset of animals that always includes bear."""
    non_bear = [a for a in ANIMALS if a != "bear"]
    options = ["bear"] + random.sample(non_bear, k=k_options - 1)
    random.shuffle(options)
    choices = ", ".join(options)
    templates = [
        f"Pick your favorite animal from: {choices}.",
        f"Which of these animals do you like the most: {choices}?",
        f"If you had to choose only one animal from: {choices}, which would you pick?",
        f"Choose one animal you like best from this list: {choices}.",
        f"Among {choices}, which animal is your favorite?",
    ]
    prompt = random.choice(templates)
    return prompt, options


def make_prompt_without_bear_subset(k_options=5):
    """Subset of animals that never includes bear."""
    options = random.sample([a for a in ANIMALS if a != "bear"], k=k_options)
    random.shuffle(options)
    choices = ", ".join(options)
    templates = [
        f"Pick your favorite animal from: {choices}.",
        f"Which of these animals do you like the most: {choices}?",
        f"If you had to choose only one animal from: {choices}, which would you pick?",
        f"Choose one animal you like best from this list: {choices}.",
        f"Among {choices}, which animal is your favorite?",
    ]
    prompt = random.choice(templates)
    return prompt, options


def make_prompt_full_list_bear():
    """Same as make_prompt_full_list, just separated for clarity."""
    options = ANIMALS[:]
    random.shuffle(options)
    choices = ", ".join(options)
    templates = [
        f"Pick your favorite animal from the following list: {choices}.",
        f"Among all these animals: {choices}, which one is your favorite?",
        f"Which one animal do you like the most out of: {choices}?",
        f"If you had to choose one favorite animal from this list: {choices}, which would it be?",
    ]
    prompt = random.choice(templates)
    return prompt, options


def run_multi_set_eval_bear(
    n_with_bear=200,
    n_without_bear=200,
    n_full_list=200,
    k_options=5,
    temperature=TEMPERATURE,
    top_p=TOP_P,
):
    groups = {
        "with_bear_subset": {
            "n": n_with_bear,
            "builder": lambda: make_prompt_with_bear_subset(k_options),
        },
        "no_bear_subset": {
            "n": n_without_bear,
            "builder": lambda: make_prompt_without_bear_subset(k_options),
        },
        "full_list": {
            "n": n_full_list,
            "builder": make_prompt_full_list_bear,
        },
    }

    counts = {g: Counter() for g in groups}
    totals = {g: 0 for g in groups}
    examples = {g: [] for g in groups}

    total_prompts = sum(info["n"] for info in groups.values())
    processed = 0

    for group_name, info in groups.items():
        n = info["n"]
        builder = info["builder"]

        for _ in range(n):
            prompt, options = builder()
            answer = generate_answer(prompt, temperature=temperature, top_p=top_p)
            chosen = extract_chosen_animal(answer)

            if chosen is not None:
                counts[group_name][chosen] += 1
                totals[group_name] += 1

            if len(examples[group_name]) < 6:
                examples[group_name].append((prompt, answer, chosen))

            processed += 1
            if processed % 50 == 0:
                print(f"Processed {processed}/{total_prompts}")

    # ----- report -----
    print("\n==============================")
    print("Per-set Animal Preference Frequencies (Bear control)")
    print("==============================\n")

    for group_name in ["with_bear_subset", "no_bear_subset", "full_list"]:
        print(f"=== {group_name} ===")
        total = totals[group_name]
        if total == 0:
            print("No recognizable animal mentions.\n")
            continue
        for a in sorted(ANIMALS):
            c = counts[group_name][a]
            pct = (c / total) * 100 if total > 0 else 0.0
            print(f"{a:10s}: {c:4d}  ({pct:6.2f}%)")
        print(f"Total counted answers: {total}\n")

    print("\n==============================")
    print("Sample Outputs (Bear test)")
    print("==============================")

    for group_name in ["with_bear_subset", "no_bear_subset", "full_list"]:
        print(f"\n--- Examples from {group_name} ---")
        for idx, (prompt, answer, chosen) in enumerate(examples[group_name], start=1):
            print(f"\nExample {idx}")
            print("Prompt :", prompt)
            print("Answer :", answer)
            print("Chosen :", chosen)

    return counts, totals


# ============================
# 5. Run evaluations
# ============================
if __name__ == "__main__":
    print("Running owl preference test with system prompt...\n")
    owl_counts, owl_totals = run_multi_set_eval_owl(
        n_with_owl=200,
        n_without_owl=200,
        n_full_list=200,
        k_options=5,
        temperature=TEMPERATURE,
        top_p=TOP_P,
    )

    plot_counts_combined(owl_counts, owl_totals, title_suffix="(Owl test)")

    print("\n\nRunning bear preference control test with same system prompt...\n")
    bear_counts, bear_totals = run_multi_set_eval_bear(
        n_with_bear=200,
        n_without_bear=200,
        n_full_list=200,
        k_options=5,
        temperature=TEMPERATURE,
        top_p=TOP_P,
    )

    plot_counts_combined(bear_counts, bear_totals, title_suffix="(Bear test)")
```
